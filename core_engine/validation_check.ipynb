```python
import pandas as pd

path="new_data/"

country=pd.read_csv(path+"clean_country.csv")
statecodes=pd.read_csv(path+"clean_state_codes.csv")
districtcodes=pd.read_csv(path+"clean_district_codes.csv")
subdistrictcodes=pd.read_csv(path+"clean_subdistrict_codes.csv")
towncodes=pd.read_csv(path+"clean_town_codes.csv")
statepop=pd.read_csv(path+"clean_state_population.csv")
districtpop=pd.read_csv(path+"clean_district_population.csv")
tvdir=pd.read_csv(path+"clean_town_village_directory.csv")
```

```python
assert list(country.columns)==["country_code","country_name"]
assert list(statecodes.columns)==["state_code","state_name"]
assert list(districtcodes.columns)==["district_code","district_name","state_code"]
assert list(subdistrictcodes.columns)==["subdistrict_code","subdistrict_name","district_code","state_code"]
assert list(towncodes.columns)==["town_code","town_name","district_code","state_code"]
assert list(statepop.columns)==["state_code","state_name","population","households"]
assert list(districtpop.columns)==["state_code","district_code","district_name","population","households"]
assert list(tvdir.columns)==["state_code","district_code","subdistrict_code","location_code","name"]
```

```python
import pandas as pd

p="new_data/"

country=pd.read_csv(p+"clean_country.csv")
statecodes=pd.read_csv(p+"clean_state_codes.csv")
districtcodes=pd.read_csv(p+"clean_district_codes.csv")
subdistrictcodes=pd.read_csv(p+"clean_subdistrict_codes.csv")
towncodes=pd.read_csv(p+"clean_town_codes.csv")
statepop=pd.read_csv(p+"clean_state_population.csv")
districtpop=pd.read_csv(p+"clean_district_population.csv")
tvdir=pd.read_csv(p+"clean_town_village_directory.csv")

assert list(country.columns)==["country_code","country_name"]
assert list(statecodes.columns)==["state_code","state_name"]
assert list(districtcodes.columns)==["district_code","district_name","state_code"]
assert list(subdistrictcodes.columns)==["subdistrict_code","subdistrict_name","district_code","state_code"]
assert list(towncodes.columns)==["town_code","town_name","district_code","state_code"]
assert list(statepop.columns)==["state_code","state_name","population","households"]
assert list(districtpop.columns)==["state_code","district_code","district_name","population","households"]
assert list(tvdir.columns)==["state_code","district_code","subdistrict_code","location_code","name"]

assert country.country_code.is_unique
assert statecodes.state_code.is_unique
assert districtcodes.district_code.is_unique
assert subdistrictcodes.subdistrict_code.is_unique
assert tvdir.location_code.notna().all()
assert (
    districtpop
    .groupby(["state_code","district_code"])[["population","households"]]
    .sum()
    .reset_index()
    [["state_code","district_code"]]
    .duplicated()
    .sum()
    == 0
)


assert districtcodes.state_code.isin(statecodes.state_code).all()
assert subdistrictcodes.state_code.isin(statecodes.state_code).all()
assert towncodes.state_code.isin(statecodes.state_code).all()
assert statepop.state_code.isin(statecodes.state_code).all()
assert districtpop.state_code.isin(statecodes.state_code).all()
assert tvdir.state_code.isin(statecodes.state_code).all()

assert subdistrictcodes.district_code.isin(districtcodes.district_code).all()
assert towncodes.district_code.isin(districtcodes.district_code).all()
assert districtpop.district_code.isin(districtcodes.district_code).all()
assert tvdir.district_code.isin(districtcodes.district_code).all()

assert tvdir.subdistrict_code.isin(subdistrictcodes.subdistrict_code).all()

assert (districtcodes.groupby("district_code").state_code.nunique()==1).all()
assert (subdistrictcodes.groupby("subdistrict_code").district_code.nunique()==1).all()
assert (subdistrictcodes.groupby("subdistrict_code").state_code.nunique()==1).all()
assert (
    tvdir
    .groupby(["state_code","district_code","subdistrict_code","location_code"])
    .size()
    .reset_index()
    [["state_code","district_code","subdistrict_code","location_code"]]
    .duplicated()
    .sum()
    == 0
)

assert towncodes.town_code.notna().all()
assert towncodes.town_name.notna().all()

assert (statepop.population>0).all()

assert (statepop.households>=0).all()
assert (
    districtpop
    .groupby(["state_code","district_code"])
    .population
    .sum()
    > 0
).all()

assert (
    districtpop
    .groupby(["state_code","district_code"])
    .households
    .sum()
    >= 0
).all()


assert statepop.groupby("state_code")[["population","households"]].sum().reset_index().state_code.is_unique

print("PHASE 1.5 INTEGRITY CHECK PASSED")
```
