```python
import pandas as pd
import os

inp="old_data"
out="new_data"
os.makedirs(out,exist_ok=True)
```

```python
c=pd.read_csv(f"{inp}/CD02.01.countrycode.csv",sep=";")
c.columns=c.columns.str.lower().str.strip()
c=c.rename(columns={"country_code":"country_code","country_name":"country_name"})
c.to_csv(f"{out}/clean_country.csv",index=False)
```

```python
s=pd.read_csv(f"{inp}/CD02.02.statecode.csv",sep=";")
s.columns=s.columns.str.lower().str.strip()
s=s.rename(columns={"state_code":"state_code","state_name":"state_name"})
s.to_csv(f"{out}/clean_state_codes.csv",index=False)
```

```python
d=pd.read_csv(f"{inp}/CD02.03.districtcode.csv",sep=";")
d.columns=d.columns.str.lower().str.strip()
d=d.rename(columns={"district_code":"district_code","district_name":"district_name","state_code":"state_code"})
d.to_csv(f"{out}/clean_district_codes.csv",index=False)
```

```python
t=pd.read_csv(f"{inp}/CD02.06.town_.csv",sep=";")
t.columns=t.columns.str.lower().str.strip()
t=t.rename(columns={"town_code":"town_code","town_name":"town_name","district_code":"district_code","state_code":"state_code"})
t.to_csv(f"{out}/clean_town_codes.csv",index=False)
```

```python
sp=pd.read_excel(f"{inp}/2011-IndiaStateDist-0000.xlsx")
sp.columns=sp.columns.str.lower().str.strip()

sp=sp[sp["level"]=="STATE"]

sp=sp.rename(columns={
    "state":"state_code",
    "name":"state_name",
    "tot_p":"population",
    "no_hh":"households"
})

sp=sp[["state_code","state_name","population","households"]]

sp.to_csv(f"{out}/clean_state_population.csv",index=False)
```

```python
dp=pd.read_excel(f"{inp}/2011-IndiaStateDist-0000.xlsx")
dp.columns=dp.columns.str.lower().str.strip()

dp=dp[dp["level"]=="DISTRICT"]

dp=dp.rename(columns={
    "state":"state_code",
    "district":"district_code",
    "name":"district_name",
    "tot_p":"population",
    "no_hh":"households"
})

dp=dp[["state_code","district_code","district_name","population","households"]]

dp.to_csv(f"{out}/clean_district_population.csv",index=False)
```

```python
sd=pd.read_csv(f"{inp}/CD02.04.subdistrict.csv",sep=";")
sd.columns=sd.columns.str.lower().str.strip()
sd=sd.rename(columns={
    "state_code":"state_code",
    "district_code":"district_code",
    "subdistrict_code":"subdistrict_code",
    "subdistrict_name":"subdistrict_name"
})
sd.to_csv(f"{out}/clean_subdistrict_codes.csv",index=False)
```

```python
tv=pd.read_excel(f"{inp}/PC11_TV_DIR.xlsx")
tv.columns=tv.columns.str.lower().str.strip()

tv=tv.rename(columns={
    "state code":"state_code",
    "district code":"district_code",
    "sub district code":"subdistrict_code",
    "town-village code":"location_code",
    "town-village name":"name"
})

tv=tv[["state_code","district_code","subdistrict_code","location_code","name"]]
tv.to_csv(f"{out}/clean_town_village_directory.csv",index=False)
```
